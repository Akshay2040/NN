```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel('Train_data.xlsx')
```

```python
df.head()
```

```python
X = df.drop('m(p)',axis=1).values
X_Train=df.drop('m(p)',axis=1).values
#Dropped the column containing the m(p), and the .values command converts the data frame to a numpy array of shape (no. of examples, no. of features)
Y = df['m(p)'].values
X = StandardScaler().fit_transform(X)#This step normalised all the input features column wise
```

```python
X
```

```python
#Step by step normalisation
s=np.squeeze(X_Train[:,0].shape)#this gives the no. of examples in a column and np.squeeze converts the array to a float data type
Mean_matrix=np.sum(X_Train, axis=0)/s #Takes the mean of each column
X_Train_mean_sub=X_Train-Mean_matrix #X-u step
Variance_matrix=np.sum(X_Train_mean_sub**2, axis=0)/(s) #Calculates the variance for each feature
X_Train_normalised=X_Train_mean_sub/np.sqrt(Variance_matrix) #gives the normalised feature matrix array
X_Train_normalised
```

# Implementation Tip
Always use mean and variance of the training set to normalise the test and dev set. Otherwise, the predictions, though will pick up the pattern but will be shifted a bit from the original values. This might lead to a negetive r2 score.
