```python
import lasio
```

```python
import lasio.examples
las = lasio.examples.open("1001178549.las")
#We can use lasio.read() to open any file file or URL.
```

```python
las.version
#The lasio.read() function returns a lasio.LASFile object. Each of the standard LAS sections can be accessed as an attribute
```

```python
lasio.SectionItems
#Each LAS section is represented as a lasio.SectionItems object. The others, for LAS 2.0 files, are present as las.well, las.curves, and las.params
```

```python
print(las.curves)
#We can print the sections as easy to read tables
```

```python
las.data.shape
```

```python
las.data 
#The data is present as a numpy.ndarray at las.data
```

```python
for curve in las.curves:
    print(curve.mnemonic + ": " + str(curve.data))
    
#We can even iterate over the curves
```

```python
las.index
#The first curve in the LAS file – usually the depth – is present as las.index
```

```python
las["IDTN"]
#Curves are also accessible from the LASFile object as items.
```
